```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Item que você deseja excluir da lista
itens_para_excluir = ['num_fotos', 'ano_de_fabricacao',  'ano_modelo', 'num_portas',  'blindado', 'entrega_delivery', 'troca' ]

# Convertendo a coluna categorical_cols em uma lista Python
numeric_cols_list = numeric_cols.tolist()

for item_para_excluir in itens_para_excluir:
  # Verifica se o item está presente na lista antes de tentar removê-lo
  if item_para_excluir in numeric_cols_list:
    numeric_cols_list.remove(item_para_excluir)

# Se você quiser, pode converter novamente para um objeto pandas.Index
numeric_cols = pd.Index(numeric_cols_list)
```

```python
import matplotlib.pyplot as plt

import seaborn as sns

# Gráfico de histogramas das variáveis numéricas
for col in numeric_cols:
  if valores_unicos[col] > 1:
    plt.figure(list(numeric_cols).index(col) + 1, figsize=(6, 3))
    sns.histplot(df[col], kde=True)
    plt.grid()
    plt.show()
```

```python
# Gráfico de boxplot das variáveis numéricas
for col in numeric_cols:
  if valores_unicos[col] > 1:
    plt.figure(list(numeric_cols).index(col) + 1, figsize = (6, 3))
    sns.boxplot(y = df[col])
    plt.grid()
    plt.show()
```

```python
categorical_cols = df.select_dtypes(include=['object', 'float', 'int', 'bool']).columns

# Item que você deseja excluir da lista
itens_para_excluir = ['modelo', 'versao', 'cidade_vendedor', 'hodometro', 'preco' ]

# Convertendo a coluna categorical_cols em uma lista Python
categorical_cols_list = categorical_cols.tolist()

for item_para_excluir in itens_para_excluir:
  # Verifica se o item está presente na lista antes de tentar removê-lo
  if item_para_excluir in categorical_cols_list:
    categorical_cols_list.remove(item_para_excluir)

# Se você quiser, pode converter novamente para um objeto pandas.Index
categorical_cols = pd.Index(categorical_cols_list)
```

```python
# Cálculo das contagens de cada categoria em cada coluna
counts_dict = {}
for col in categorical_cols:
  counts_dict[col] = df[col].value_counts()
```

```python
# Ordenar as colunas pelo número de frequências (contagens)
sorted_cols = sorted(counts_dict, key=lambda col: counts_dict[col].sum(), reverse=True)
```

```python
# Plotar o gráfico de barras para cada coluna, em ordem decrescente
for col in sorted_cols:
  if valores_unicos[col] > 1 and col != 'id':
    plt.figure(list(categorical_cols).index(col) + 1, figsize=(6, 6))
    sns.countplot(y=col, data=df, order=df[col].value_counts().index)
    plt.xticks(rotation=45)
    plt.grid()
    plt.show()
```

# 4. Análise Exploratória de Dados (EDA) e Hipóteses de Negócio

### Hipóteses de negócio criadas

```python
# 1. Hipótese de Negócio: Qual a cidade com o menor preço médio de venda de um carro do tipo sedã com transmissão automática?
seda_automatico = df[(df['tipo'] == 'Sedã') & (df['cambio'] == 'Automática')]
cidade  = seda_automatico.groupby('cidade_vendedor')['preco'].mean().idxmin()
cidade
```

NOTA: Localizado no Estado de São Paulo.

```python
# 2. Hipótese de Negócio: Qual a cidade com o maior preço médio de venda entre os veículos anunciados?
cidade = df.groupby('cidade_vendedor')['preco'].mean().idxmax()
cidade
```

NOTA: Localizado no Estado de Goiás.

```python
# 3. Hipótese de Negócio: Qual a cidade com o menor preço médio de venda entre os veículos anunciados da marca FORD, modelo FIESTA?
fiesta = df[(df['modelo'] == 'FIESTA')]
cidade = fiesta.groupby('cidade_vendedor')['preco'].mean().idxmin()
cidade
```

NOTA: Localizado no Estado de São Paulo.

### Hipóteses de negócio respondidas

```python
# 1. Qual o melhor estado cadastrado para se vender um carro de marca popular e por quê?
popular  = df[df['marca'] == 'VOLKSWAGEN']
estado = popular.groupby('estado_vendedor')['preco'].mean().idxmax()
estado
```

```python
# 2. Qual o melhor estado para se comprar uma picape com transmissão automática e por quê?
picape_automatica = df[(df['tipo'] == 'Picape') & (df['cambio'] == 'Automática')]
estado  = picape_automatica.groupby('estado_vendedor')['preco'].mean().idxmin()
estado
```

```python
# 3. Qual o melhor estado para se comprar carros que ainda estejam dentro da garantia de fábrica e por quê?
carro_na_garantia = df[df['garantia_de_fábrica'] == 'Garantia de fábrica']
estado = carro_na_garantia.groupby('estado_vendedor')['preco'].mean().idxmin()
estado
```
